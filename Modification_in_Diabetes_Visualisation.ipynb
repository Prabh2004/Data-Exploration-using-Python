```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

# Data Visualisation Using Graphical Representation using Various Graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame
# For simplicity, let's consider a dst with columns 'Age' and 'Pregnancies'

# Scatter plot
sns.scatterplot(x='Age', y='BMI', data= df)
plt.title('Scatter Plot of Age vs BMI')
plt.show()

# Histogram
sns.histplot(data=df, x='Age', bins=20, color='blue', alpha=0.9, label='Age')
sns.histplot(data=df, x='BMI', bins=20, color='red', alpha=0.9, label='AGe')
plt.title('Histogram of Age and BMI')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Box plot
sns.boxplot(data=df[['Age', 'Pregnancies']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=df[['PG Concentration','Diastolic BP']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=df[['Serum Ins','BMI','DP Function']])
plt.title('Box Plot of Age and feature2')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='class', data=df)
plt.title('Distribution of Classes')
plt.show()
```

## Corelation Matrix

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Pairplot Graph repesent

this graph display scattering the numeric scattering between classes


```python
selected_features = ['BMI', 'DP Function', 'Pregnancies', 'class']
sns.pairplot(df[selected_features], hue='class', markers=['o', 's'])
plt.suptitle('Pairplot of Features', y=1.00, ha='center', fontsize='large')
plt.show()
```

```python
X = df.drop('class', axis=1)
y = df['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='bar')
plt.title('Feature Importances')
plt.show()
```
